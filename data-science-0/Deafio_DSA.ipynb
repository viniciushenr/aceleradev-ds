```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA = os.path.abspath('.')
```

```python
#Dados disponibilizados pelas Kaggle, baixe e insira na mesma pasta que este .ipynb file
# https://www.kaggle.com/orgesleka/used-cars-database
```

```python
df = pd.read_csv(os.path.join(DATA,'autos.csv'),sep=',',encoding='ISO-8859-1')
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
fig, ax = plt.subplots(figsize=(10,10))
sns.distplot(pd.DataFrame(df['yearOfRegistration']).query("yearOfRegistration>1900 & yearOfRegistration<2020"))
ax.set_title('Distribuição de Veículos com base no Ano de Registro', fontsize= 15)
plt.ylabel("Densidade (KDE)", fontsize= 15)
plt.xlabel("Ano de Registro", fontsize= 15)
plt.show()
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
fig, ax = plt.subplots(figsize=(10,10))
sns.boxplot(x='vehicleType',y='price',data=df.dropna().query("price<30000"))
ax.set_title('Boxplot para avaliar os outliers', fontsize= 15)
plt.ylabel("Preço", fontsize= 15)
plt.xlabel("Tipo de veículos", fontsize= 15)
plt.show()
```

## Contagem total de veículos à venda conforme o tipo de veículo¶

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 
fig, ax = plt.subplots(figsize=(10,10))
sns.countplot(x='vehicleType',data=pd.DataFrame(df['vehicleType'].dropna()))
ax.set_title('Número de veículos pertencentes a cada categoria', fontsize= 15)
plt.ylabel("Quantidade", fontsize= 15)
plt.xlabel("Tipo de veículos", fontsize= 15)
plt.show()
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
fig, ax = plt.subplots(figsize=(10,10))
sns.countplot(y='brand',data=pd.DataFrame(df['brand'].dropna()))
ax.set_title('Número de veículos pertencentes a cada marca', fontsize= 15)
plt.ylabel("Marcas", fontsize= 15)
plt.xlabel("Quantidade", fontsize= 15)
plt.show()
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(10,10))
sns.barplot(x="vehicleType", y="price",hue="gearbox", data=df.dropna().query("price<30000"))
ax.set_title('Preço médio dos veículos com base no tipo de veículo e tipo de caixa de câmbio', fontsize= 15)
plt.ylabel("Preço médio", fontsize= 15)
plt.xlabel("Tipo de veículos", fontsize= 15)
plt.show()
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(10,10))
sns.barplot(x="fuelType", y="price",hue="gearbox", data=df.dropna().query("price<30000"))
ax.set_title('Preço médio dos veículos com base no tipo de combustível e tipo de caixa de câmbio', fontsize= 15)
plt.ylabel("Preço médio", fontsize= 15)
plt.xlabel("Tipo de combustível", fontsize= 15)
plt.show()
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
fig, ax = plt.subplots(figsize=(10,10))
sns.barplot(x="vehicleType", y="powerPS",hue="gearbox", data=df.dropna())
ax.set_title('Potência média dos veículos com base no tipo de veículo e tipo de caixa de câmbio', fontsize= 15)
plt.ylabel("Potência média", fontsize= 15)
plt.xlabel("Tipo de veículos", fontsize= 15)
plt.show()
```
